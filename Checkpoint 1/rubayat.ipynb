```python
import pandas as pd

merged_data = pd.read_csv('merged_data.csv')
```

```python
print(merged_data.head())
print(merged_data.info())
```

```python
statistics = merged_data.describe(include='all')
print(statistics)
```

```python
correlation_matrix = merged_data.corr()
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(merged_data['C_we'], bins=30, kde=True)
plt.title('Distribution of Computer Usage on Weekends')
plt.xlabel('Hours per Day')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.boxplot(x='gender', y='C_we', data=merged_data)
plt.title('Computer Usage on Weekends by Gender')
plt.xlabel('Gender')
plt.ylabel('Hours per Day')
plt.show()
```
